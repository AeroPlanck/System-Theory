# 分岔图分析本Notebook实现了一个微分方程 $f'(t) = (f(t)^2 + k)f(t)$ 的分岔图分析。通过改变参数$k$的值，我们可以观察系统的稳态行为如何变化，并找出可能的分岔点。

## 导入必要的库首先导入我们需要的Python库：- NumPy：用于数值计算- Matplotlib：用于绘图- SciPy的solve_ivp：用于求解微分方程- multiprocessing.Pool：用于并行计算- time：用于计时

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
from multiprocessing import Pool
import time
```

## 定义微分方程我们研究的微分方程为：$f'(t) = (f(t)^2 + k)f(t)$，其中$k$是控制参数。

```python
def f_prime(t, f, k):
    """
    定义微分方程 f'(t) = (f(t)^2 + k)f(t)
    
    参数:
        t: 时间变量
        f: 函数值 f(t)
        k: 参数 k
    
    返回:
        f'(t) 的值
    """
    return (f**2 + k) * f
```

## 分岔图计算类下面定义一个类来计算和绘制分岔图。这个类包含以下主要方法：- `__init__`：初始化参数- `solve_equation`：求解给定k值的微分方程- `compute_bifurcation`：计算分岔图数据- `plot_bifurcation`：绘制分岔图- `analyze_stability`：分析系统稳定性，找出分岔点

```python
class BifurcationDiagram:
    def __init__(self, k_range=(-2, 2), k_points=200, t_max=20, initial_value=0.01):
        """
        初始化分岔图计算类
        
        参数:
            k_range: 参数k的范围，元组 (k_min, k_max)
            k_points: k值的采样点数
            t_max: 时间积分上限
            initial_value: 初始值 f(0)
        """
        self.k_min, self.k_max = k_range
        self.k_points = k_points
        self.k_values = np.linspace(self.k_min, self.k_max, self.k_points)
        self.t_max = t_max
        self.initial_value = initial_value
        self.results = []
    
    def solve_equation(self, k):
        """
        求解给定k值的微分方程
        
        参数:
            k: 参数值
            
        返回:
            k值和最终的f值
        """
        # 定义微分方程右侧函数
        def equation(t, y):
            return f_prime(t, y, k)
        
        # 使用Runge-Kutta方法求解微分方程
        # 注意：对于某些k值，解可能会发散到无穷大，因此设置一个事件函数来提前终止
        def event(t, y):
            return 1e6 - abs(y)  # 当|y|达到1e6时停止积分
        event.terminal = True
        
        # 求解微分方程
        solution = solve_ivp(
            equation, 
            [0, self.t_max], 
            [self.initial_value], 
            method='RK45',
            events=event,
            dense_output=True
        )
        
        # 获取最终时刻的值
        if solution.t[-1] < self.t_max * 0.9:  # 如果提前终止，说明解发散了
            final_value = np.nan  # 标记为NaN
        else:
            final_value = solution.y[0, -1]
            
        return k, final_value
    
    def compute_bifurcation(self, use_multiprocessing=True, n_processes=4):
        """
        计算分岔图数据
        
        参数:
            use_multiprocessing: 是否使用多进程加速计算
            n_processes: 进程数
        """
        print(f"开始计算分岔图，参数k范围: [{self.k_min}, {self.k_max}]，采样点数: {self.k_points}")
        start_time = time.time()
        
        if use_multiprocessing:
            # 使用多进程加速计算
            with Pool(n_processes) as pool:
                self.results = pool.map(self.solve_equation, self.k_values)
        else:
            # 单进程计算
            self.results = [self.solve_equation(k) for k in self.k_values]
        
        # 将结果转换为数组形式，便于绘图
        self.k_array = np.array([res[0] for res in self.results])
        self.f_array = np.array([res[1] for res in self.results])
        
        end_time = time.time()
        print(f"计算完成，耗时: {end_time - start_time:.2f}秒")
    
    def plot_bifurcation(self, save_path=None):
        """
        绘制分岔图
        
        参数:
            save_path: 保存图像的路径，如果为None则不保存
        """
        plt.figure(figsize=(10, 6))
        
        # 绘制分岔图
        plt.scatter(self.k_array, self.f_array, s=1, c='blue', alpha=0.7)
        
        # 添加水平线 f=0
        plt.axhline(y=0, color='r', linestyle='--', alpha=0.5)
        
        # 设置图表标题和标签
        plt.title("分岔图: $f'(t) = (f(t)^2 + k)f(t)$", fontsize=14)
        plt.xlabel("参数 $k$", fontsize=12)
        plt.ylabel("稳态值 $f(\\infty)$", fontsize=12)
        
        # 设置坐标轴范围
        plt.xlim(self.k_min, self.k_max)
        valid_values = self.f_array[~np.isnan(self.f_array)]
        if len(valid_values) > 0:
            y_min, y_max = np.min(valid_values), np.max(valid_values)
            y_range = y_max - y_min
            plt.ylim(y_min - 0.1 * y_range, y_max + 0.1 * y_range)
        
        # 添加网格线
        plt.grid(True, alpha=0.3)
        
        # 保存图像
        if save_path:
            plt.savefig(save_path, dpi=300, bbox_inches='tight')
            print(f"图像已保存至: {save_path}")
        
        plt.tight_layout()
        plt.show()
    
    def analyze_stability(self):
        """
        分析系统的稳定性，找出分岔点
        """
        # 找出f值从正变为负或从负变为正的k值，这些可能是分岔点
        bifurcation_points = []
        for i in range(1, len(self.f_array)):
            if (self.f_array[i-1] * self.f_array[i] < 0) or \
               (np.isnan(self.f_array[i-1]) and not np.isnan(self.f_array[i])) or \
               (not np.isnan(self.f_array[i-1]) and np.isnan(self.f_array[i])):
                bifurcation_points.append(self.k_array[i])
        
        if bifurcation_points:
            print("\n可能的分岔点:")
            for point in bifurcation_points:
                print(f"k = {point:.4f}")
        else:
            print("\n在给定范围内未找到明显的分岔点")
```

## 运行分析现在我们可以创建一个分岔图对象，并运行分析。我们将：1. 设置参数范围和初始条件2. 计算分岔图数据3. 绘制分岔图4. 分析系统稳定性

```python
# 创建分岔图对象
bifurcation = BifurcationDiagram(
    k_range=(-2, 2),  # 参数k的范围
    k_points=500,     # k值的采样点数
    t_max=50,         # 时间积分上限
    initial_value=0.01  # 初始值，设为一个小的非零值
)

# 计算分岔图数据
bifurcation.compute_bifurcation(use_multiprocessing=True, n_processes=4)
```

```python
# 绘制分岔图
bifurcation.plot_bifurcation(save_path="bifurcation_diagram.png")
```

```python
# 分析稳定性
bifurcation.analyze_stability()
```

## 结论通过分岔图分析，我们可以观察到系统在不同参数$k$值下的稳态行为。分岔点表示系统稳态行为发生质变的临界点，这对于理解系统的动力学特性非常重要。在这个特定的微分方程中，我们可以看到：- 当$k < 0$时，系统可能有多个稳态解- 当$k = 0$时，系统处于临界状态- 当$k > 0$时，系统的行为会发生显著变化这种分析方法可以应用于各种非线性动力系统的研究。
